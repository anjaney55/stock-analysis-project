```python
import yfinance as yf

stock = yf.download("AAPL", period="5d", interval="1h")
stock.to_csv("stock_data.csv")

stock.head()
```

```python
import requests

url = "https://newsapi.org/v2/everything?q=apple&apiKey=ec430183bc9f43968b1ef3f5cbbacb12"

response = requests.get(url)
data = response.json()

articles = []

for article in data["articles"]:
    articles.append(article["title"])

import pandas as pd
news_df = pd.DataFrame(articles, columns=["title"])

news_df.to_csv("news_data.csv", index=False)
```

```python
#I used financial market data and real-time news to analyze how sentiment impacts stock movement.
```

```python
news_df=pd.read_csv("news_data.csv")
stock_df=pd.read_csv("stock_data.csv")
```

```python
#CLEAN NEWS DATA
```

```python
#Remove Nulls
news_df.dropna(inplace=True)
```

```python
#Remove Duplicates
news_df.drop_duplicates(inplace=True)
```

```python
import re
def clean_text(text):
    text=text.lower()  #lowercase
    text = re.sub(r"http\S+", "", text)  # remove links
    text = re.sub(r"[^a-zA-Z ]", "", text)  # remove symbols
    return text

news_df["clean_title"] = news_df["title"].apply(clean_text)
```

```python
#CLEAN STOCK DATA
stock_df.columns
```

```python
#Reset Index
stock_df.reset_index(inplace=True)
```

```python
#Rename Columns
stock_df.rename(columns={"index": "time"}, inplace=True)
```
