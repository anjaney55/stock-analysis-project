```python
import requests

url = (
    "https://newsapi.org/v2/everything?"
    "q=Apple&"
    "sortBy=publishedAt&"
    "language=en&"
    "pageSize=100&"
    "apiKey=ec430183bc9f43968b1ef3f5cbbacb12"
)

response = requests.get(url)

data = response.json()

articles = []

for article in data ["articles"]:
    articles.append([
        article["title"],
        article["publishedAt"],
        article["source"]["name"]
            ])

news_df = pd.DataFrame(
    articles,
    columns=[
        "title",
        "publishedAt",
        "source"
    ])

news_df.to_csv(
    "news_data_fresh.csv",
    index=False
)

news_df.head()
    
```

```python
#CLEAN NEWS DATA
news_df.columns
```

```python
#Remove Nulls
news_df.dropna(inplace=True)
```

```python
#Remove Duplicates
news_df.drop_duplicates(inplace=True)
```

```python
import re
def clean_text(text):
    text=text.lower()  #lowercase
    text = re.sub(r"http\S+", "", text)  # remove links
    text = re.sub(r"[^a-zA-Z ]", "", text)  # remove symbols
    return text

news_df["clean_title"] = news_df["title"].apply(clean_text)
```

```python
#CLEAN STOCK DATA
stock_df.columns
```

```python
#Reset Index
stock_df.reset_index(inplace=True)
```

```python
#Rename Columns
stock_df.rename(columns={"index": "time"}, inplace=True)
```

```python
stock_df.columns
```

```python
#Select Important Columns
stock_df = stock_df[["time","Open","Close"]]
```

```python
'''Understand WHY Sentiment?
Think like business:
👉 News says:
“Apple stock crashes” → Negative
“Apple beats expectations” → Positive
👉 Company wants:
“Can we predict stock movement from news?'''
news_df.columns
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

news_df["sentiment"] = news_df["clean_title"].apply(get_sentiment)
```

```python
news_df.columns
```

```python
news_df = news_df[["title","clean_title", "time", "sentiment",]]
```

```python
#INSERT SENTIMENT INTO SQL
```

```python
def sentiment_label(score):
    if score>0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

news_df["sentiment_label"] = news_df["sentiment"].apply(sentiment_label)
    
```
