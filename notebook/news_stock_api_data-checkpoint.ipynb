```python
news_df=pd.read_csv("news_data.csv")
stock_df=pd.read_csv("stock_data.csv")
```

```python
#CLEAN NEWS DATA
```

```python
#Remove Nulls
news_df.dropna(inplace=True)
```

```python
#Remove Duplicates
news_df.drop_duplicates(inplace=True)
```

```python
import re
def clean_text(text):
    text=text.lower()  #lowercase
    text = re.sub(r"http\S+", "", text)  # remove links
    text = re.sub(r"[^a-zA-Z ]", "", text)  # remove symbols
    return text

news_df["clean_title"] = news_df["title"].apply(clean_text)
```

```python
#CLEAN STOCK DATA
stock_df.columns
```

```python
#Reset Index
stock_df.reset_index(inplace=True)
```

```python
#Rename Columns
stock_df.rename(columns={"index": "time"}, inplace=True)
```

```python
stock_df.columns
```

```python
#Select Important Columns
stock_df = stock_df[["time","Open","Close"]]
```

```python
'''Understand WHY Sentiment?
Think like business:
👉 News says:
“Apple stock crashes” → Negative
“Apple beats expectations” → Positive
👉 Company wants:
“Can we predict stock movement from news?'''
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

news_df["sentiment"] = news_df["clean_title"].apply(get_sentiment)
```

```python
news_df.head()
```

```python
news_df = news_df[["clean_title", "sentiment"]]
```

```python
news_df.head()
```

```python
print(news_df.columns)
```

```python
#INSERT SENTIMENT INTO SQL
```

```python
def sentiment_label(score):
    if score>0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

news_df["sentiment_label"] = news_df["sentiment"].apply(sentiment_label)
    
```

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="market_analysis"
)

cursor = conn.cursor()

for i, row in news_df.iterrows():
    cursor.execute(
        "INSERT INTO news (clean_title, sentiment, sentiment_label) VALUES (%s, %s, %s)",
        (row["clean_title"], row["sentiment"], row["sentiment_label"])
    )

conn.commit()
```
