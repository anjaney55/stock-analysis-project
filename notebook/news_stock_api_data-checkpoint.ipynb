```python
import yfinance as yf

stock = yf.download("AAPL", period="5d", interval="1h")
stock.to_csv("stock_data.csv")

stock.head()
```

```python
import requests

url = "https://newsapi.org/v2/everything?q=apple&apiKey=ec430183bc9f43968b1ef3f5cbbacb12"

response = requests.get(url)
data = response.json()

articles = []

for article in data["articles"]:
    articles.append(article["title"])

import pandas as pd
news_df = pd.DataFrame(articles, columns=["title"])

news_df.to_csv("news_data.csv", index=False)
```
